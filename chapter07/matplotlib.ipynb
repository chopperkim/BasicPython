```python
# 그래프(Graph: 자료구조에서 관계를 표현할 때 가장 많이 사용하는 것 중 하나)
# 사람과의 관계, 데이터의 흐름(방향 그래프), 개체 사이의 관계, 네트워크

import networkx as nx
graph = nx.Graph() # new Graph()랑 같은 코드
graph.add_nodes_from(['A', 'B', 'C', 'D', 'E'])
graph.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('A', 'E'), ('A', 'D')])

nx.draw(graph, with_labels=True)
```

```python
graph1 = nx.Graph()
# 노드 추가 및 삭제
graph1.add_node('David')
graph1.add_node('John')
graph1.add_node('Mary')
graph1.add_node('Julia')
graph1.add_node('Juda')
graph1.add_nodes_from(['Judy', 'Karen'])
graph1.remove_node('Julia')

# 엣지 추가 및 제거
graph1.add_edge('David', 'John')
graph1.add_edge('David', 'Mary')
graph1.add_edge('David', 'Karen')
graph1.add_edges_from([('John', 'Mary'), ('John', 'Judy')])
graph1.remove_edge('Mary', 'John')
nx.draw(graph1, with_labels=True, font_weight='bold', nodelist=dict(graph1.degree).keys(), node_size=[v * 1000 for v in dict(graph1.degree).values()])
graph1.nodes
```

```python
graph3 = nx.DiGraph()
graph3.add_weighted_edges_from([(1, 2, 10), (2, 3, 20)])
graph3.add_edge(1, 3, weight=30)
pos = nx.spring_layout(graph3)

nx.draw(graph3, pos=pos, with_labels=True)
labels = nx.get_edge_attributes(graph3, 'weight')
nx.draw_networkx_edge_labels(graph3, pos, edge_labels=labels)
```

```python
df = pd.DataFrame({'from': ['A', 'B', 'C', 'A', 'E', 'D'], 'to': ['D', 'A', 'E', 'C', 'A', 'E'], 'weight': [1, 2, 3, 4, 5, 6]})
graph4 = nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.DiGraph())
nx.draw(graph4, with_labels=True)
```

```python
graph5 = nx.Graph()
graph5.add_nodes_from(['A', 'B', 'C', 'D', 'E'])
graph5.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E')])
nx.draw(graph5, with_labels=True)
```
